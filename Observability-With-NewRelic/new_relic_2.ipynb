# BROWSER MONITORING

Browser monitoring in New Relic provides a powerful solution for optimizing web application performance and user experience. By installing a lightweight JavaScript agent, developers gain invaluable insights into various key performance indicators, including:

- **Largest Contentful Paint (LCP):** LCP measures the time it takes for the largest content element on a web page to become visible to the user. It is a crucial metric for assessing page load speed and user perception of performance.

- **First Input Delay (FID):** FID quantifies the responsiveness of a web page by measuring the time between a user's first interaction (e.g., clicking a button or link) and the browser's response. A low FID indicates a more responsive and interactive user experience.

- **Cumulative Layout Shift (CLS):** CLS measures the visual stability of a web page by quantifying the amount of unexpected layout shifts that occur during the page load process. It indicates how much content elements move around the viewport, potentially causing a disruptive user experience. Minimizing CLS helps ensure that users can interact with the page without accidentally clicking on unintended elements due to layout shifts.

With these parameters, including Largest Contentful Paint (LCP), First Input Delay (FID), and Cumulative Layout Shift (CLS), developers gain a comprehensive understanding of their web application's performance and user experience. By monitoring and optimizing these metrics through New Relic's browser monitoring capabilities, organizations can deliver fast, responsive, and visually stable web experiences to their users, ultimately improving user satisfaction and engagement.

**AWS RESOURCES INTERGRATION NEW RELIC:**

Integration of AWS resources with New Relic facilitates seamless monitoring and observability of cloud infrastructure. Here's a step-by-step guide on how to set up this connection:

1. **Access Infrastructure in New Relic UI:** Navigate to the Infrastructure section within the New Relic UI.

2. **Connect to AWS Account:** Choose the option to integrate with AWS. You'll have two options: metric streams or API polling.

3. **Configure AWS Console:** In the AWS console, create a new IAM role with the appropriate permissions. Ensure the trusted entity type is set to AWS account, and specify the Account ID and External ID provided by New Relic.

4. **Assign Permissions:** Grant the IAM role read-only access to your AWS resources.

5. **Copy ARN:** Once the IAM role is created, copy its Amazon Resource Name (ARN).

6. **Add ARN in New Relic:** Paste the ARN along with your AWS account name in New Relic. Follow the provided CloudFormation script to create resources.

7. **License Key Integration:** During the setup process, provide your New Relic license key when prompted.

8. **Wait for Stack Creation:** Allow time for the CloudFormation script to create the necessary resources.

9. **Access AWS Metrics:** Once the stack creation is complete, metrics from AWS will begin to populate in New Relic. You'll have visibility into various AWS services, including Lambda functions.

- **Monitor AWS Resources:** Utilize New Relic's monitoring capabilities to track and analyze AWS resources, gaining valuable insights into performance and health metrics.



# Synthetic Monitorying:

4. **Custom Execution Frequency:** Set the execution frequency of synthetic monitors according to your needs. Regular checks can provide continuous assurance of service availability and performance, helping to identify and address issues proactively.

By leveraging synthetic monitors for pre-instrumentation checks and ongoing monitoring, organizations can detect and resolve issues before they impact users, ensuring a seamless and reliable user experience.

**New Relic Deployments, CICD Integration:**

In New Relic's CI/CD framework, the Deployments feature establishes a critical three-hour change tracking window surrounding deployment times, essential for detecting anomalies. To ensure smooth integration, users must create a dedicated profile specifically for deployments, following New Relic's guided configuration. Executing provided commands on the deployment build's host facilitates seamless integration. Subsequently, the change tracking submenu in New Relic's interface conveniently catalogs all deployment records, enabling users to monitor changes effectively.

Integrating APM and Services with your CI/CD pipeline using your favorite CI/CD tool streamlines deployment processes and enhances visibility into changes. Here's how to leverage this integration effectively:

1. **Deployment Summary in APM & Services:** Begin by configuring your CI/CD tool to provide a deployment summary within the APM & Services dashboard. This summary should include key details such as deployment time, version, and associated changes.

2. **Tracking Changes Before and After Deployment:** Ensure that your CI/CD tool tracks changes occurring three hours before and after each deployment. This window allows for comprehensive monitoring of the deployment process and its impact on system performance.

3. **Initiating Change Tracking Process:** Once a deployment is triggered, initiate the change tracking process within your CI/CD tool. This process should capture all modifications made to the codebase, configuration, or infrastructure during the deployment.

4. **Dotted Time for Change Creation:** Each change created during the deployment process should be timestamped with dotted time, indicating when the modification occurred. This granularity enables precise tracking of changes and their effects on system behavior.

5. **Tracking Last Deployment:** Ensure that the CI/CD tool tracks the details of the last deployment, including version numbers, timestamps, and associated changes. This information serves as a reference point for future deployments and troubleshooting efforts.

By integrating APM and Services with your CI/CD pipeline and following these steps, you can effectively monitor deployments, track changes, and ensure the stability and performance of your applications and services throughout the software development lifecycle.

# Working with Issues & Alerts

Managing issues and alerts effectively is crucial for maintaining the stability and performance of your applications. Here's how you can create alerts and configure notification channels using New Relic:

1. **Define Alert Conditions:** Begin by defining alert conditions based on metrics, synthetic monitors, or custom NRQL queries. You can set thresholds using Signal Threshold or Droop Alert Conditions to trigger alerts when certain criteria are met.

2. **Raise Incidents:** When an alert condition is breached, incidents are automatically raised. These incidents represent issues that need attention.

3. **Group Incidents into Decisions:** Incidents can be grouped into decisions, which help organize and prioritize issues for resolution.

4. **Configure Notification Channels:** Use workflows to decide where and when notifications should be sent. Set up destinations for notifications, such as email, Slack, or other channels, to ensure that the right stakeholders are notified promptly.

5. **Leverage Applied Intelligence:** New Relic's Applied Intelligence feature helps anticipate and prevent issues by analyzing patterns and trends in your data. It keeps you ahead of potential issues before they impact your application.

6. **Create Alert Policies:** In the New Relic UI, navigate to alert policies and create a new policy. Name the policy according to your classification and define the conditions for triggering alerts. You can choose to create one issue per condition to ensure clarity and focus.

7. **Customize Alert Conditions:** Customize alert conditions based on your specific requirements. You can use predefined metrics like APM key golden metrics or APM service granular metrics, or create custom conditions tailored to your application's needs.

8. **Link Runbooks for Documentation:** For ease of troubleshooting, you can link runbooks or documentation (e.g., Confluence) to alert conditions. This provides additional context and guidance for engineers tasked with resolving issues.

By following these steps and leveraging New Relic's alerting and notification features, you can proactively monitor your applications, quickly identify and address issues, and ensure optimal performance and reliability.

# NRQL

NRQL (New Relic Query Language) is a powerful query language used to retrieve and analyze data collected by New Relic. It provides a flexible and intuitive way to query metrics, events, and logs, enabling users to gain valuable insights into their application's performance and behavior. NRQL queries can be used to filter, aggregate, and visualize data, allowing users to answer specific questions about their systems. With NRQL, you can create custom dashboards, alerts, and reports tailored to your unique monitoring needs. Its syntax is similar to SQL, making it accessible to users familiar with database querying languages. Overall, NRQL is a fundamental tool for extracting actionable insights from your New Relic data and optimizing your application's performance.

````python
Navigate to Left toolbar -> Query Your data -> Query Builder -> nice layout to play with queries
````

nrqlused.png

````python
SELECT * FROM Transaction
````

Above NRQL query retrieves all data from the Transaction event type, providing a comprehensive view of transaction-related information within your application.

````python
SELECT average(duration) FROM Transaction
````

Above NRQL query calculates the average duration of transactions in your application, providing insights into the typical time it takes for transactions to complete.

````python
SELECT average(duration) FROM Transaction WHERE transactionType in ('web', 'Web') AND transactionSubType = 'Function'
# we can customize as per need
````

Above NRQL query calculates the average duration of web transactions categorized as "Function" within your application. It specifically focuses on web transactions that involve executing functions, providing insights into the performance of these specific operations.

````python
SELECT average(duration), max(duration) FROM Transaction FACET host WHERE transactionType in ('web', 'Web')
# we don't have group by in NRQL we have FACET which comes always before where statement, if where exists
````

Above NRQL query calculates the average and maximum duration of web transactions, grouped by host. It provides insights into the performance of web transactions across different hosts within your application infrastructure.

````python
SELECT average(duration), count(duration) FROM Transaction FACET host WHERE transactionType in ('web', 'Web')
````

Above NRQL query calculates the average duration and counts the occurrences of web transactions, grouped by host. It provides insights into both the average response time and the frequency of web transactions across different hosts within your application infrastructure.

````python
#We can apply time limits

SELECT average(duration) FROM Transaction FACET host WHERE transactionType in ('web','Web') SINCE 4 hours ago

SELECT average(duration) FROM Transaction FACET host WHERE transactionType in ('web','Web') SINCE 1 hour ago UNTIL 1 minute ago
````

- The first NRQL query calculates the average duration of web transactions, grouped by host, within the last 4 hours. It provides insights into the performance of web transactions across different hosts over this time period.

- Second NRQL query calculates the average duration of web transactions, grouped by host, within the last hour up until one minute ago. It offers a more granular analysis of recent web transaction performance within your application infrastructure.

````python
# Plot it using TIMESERIES, customizable, find all options on the right hand side

SELECT average(duration), average(priority) FROM Transaction FACET appName TIMESERIES
````

Above NRQL query calculates the average duration and average priority of transactions over time, grouped by application name. The "TIMESERIES" clause ensures that the results are plotted over time, providing insights into how transaction duration and priority fluctuate across different applications.

````python
SELECT uniques(dataType) AS 'metric' FROM Metric
````

Above NRQL query retrieves unique values from the "dataType" attribute in the Metric event type, representing different types of metrics collected by New Relic. It provides insights into the variety of metrics being tracked within your application or system.

````python
SELECT * FROM Metric WHERE metricName Like 'newrelic%' LIMIT 10

SELECT average(apm.service.overview.web) FROM Metric WHERE appName='BookStore Flask App'
````

1. This retrieves data from the Metric event type where the metricName begins with 'newrelic'. It limits the results to 10 records.

2. This query calculates the average value of the 'apm.service.overview.web' metric for the 'BookStore Flask App' application. It provides insights into the average performance of web transactions within the specified application.

**Service Levels Explained:**

In the realm of performance monitoring, service levels play a pivotal role in ensuring the quality of service meets predefined standards. Here's a breakdown:

1. **SLA (Service Level Agreement):** SLA is a formal agreement between a service provider and its customers, outlining the level of service expected. It typically includes metrics such as uptime, response time, and availability, along with consequences for failing to meet these metrics.

2. **SLO (Service Level Objective):** SLOs are specific, measurable targets derived from SLAs. They represent the performance goals that must be achieved to fulfill the terms of the SLA. For example, an SLO might specify that the system should maintain an uptime of 99.9% or respond to requests within 200 milliseconds.

3. **Apm & Services Monitoring:** Within New Relic's APM & Services monitoring platform, service levels are crucial for evaluating performance against predefined benchmarks. Users can establish baselines for key metrics such as latency and success rates, allowing them to track performance over time and ensure that SLOs are met.

By leveraging service levels, organizations can set clear performance expectations, monitor adherence to these standards, and take proactive measures to maintain optimal service quality for their customers.
